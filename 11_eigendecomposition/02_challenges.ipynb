## 01: Eigenvalues of Diagonal and Triangular matrices

1. Create diagonal matrix and extract eigenvalues
2. Repeat but for L and U matrices

```python
import numpy as np
from scipy import linalg as la
import sympy
```

```python
# Create diagonal matrix, 2x2
diag = np.random.randint(0, high=100, size=2)
D = np.zeros(shape=(2,2))

for i in range(len(diag)):
    D[i, i] = diag[i]

D
```

```python
# Find eigenvalues
e_val, e_vec =  la.eig(D)

e_val
```

Interestingly, we can see that the eigenvalues of a diagonal matrix are the values on the diagonal of a matrix

```python
e_vec
```

What's even more interesting is that the eigenvectors create the identity matrix!

This is probably because a diagonal matrix is a simply stretch, and so the eigenvectors dont create any rotation. Makes sense.

```python
# Create L and U matrices, 2x2
A = np.random.random(size=(2, 2))

L, U = la.lu(A, permute_l  = True)

U
```

```python
# Find eigenvalues
e_val, e_vec =  la.eig(U)

e_val
```

We can see that this is also true for an upper triangular matrix

```python
e_vec
```

The difference between an upper triangular matrix and a diagonal matrix is that there is a combination of stretching and some rotating where the number of stretching components is equal to 1.

## 02: Eigenvalues of random matrices

1. Generate large matrix of random values
2. Extract its eigenvalues

Plot real part by the imaginary part.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("seaborn")

# Create random 40x40 matrix
for i in range(300):
    B = np.random.rand(40, 40)
    e_val, e_vec = la.eig(B)
    e_val = e_val[e_val.real <= 15]
    plt.scatter(e_val.real, e_val.imag, alpha=0.1)
plt.show()
```

We can see that the random eigenvalues create a circle when done many times with a line going across the center. This is where the eigenvalues are strictly imaginary.

The radius of this circle is relative to the size of Matrix A.

## 03: Trace and Determinant, eigenvalues sum and product

The trace of a square matrix is equal to the sum of its eigenvalues
The determinant of a matrix is equal to the product of its eigenvalues

1. Create square matrix
2. Show the above

```python
C = np.random.randint(0, 100, size = (3, 3))

C
```

```python
C_trace = C.trace()
C_det = np.linalg.det(C)

C_trace, C_det
```

```python
C_eigval, C_eigvec = la.eig(C)
```

```python
# Is trace equal to sum of eigvals?
C_eigval.real.sum(), C_trace
```

```python
# Is det equal to the product of eigvals?
C_eigval.real.prod(), C_det
```
