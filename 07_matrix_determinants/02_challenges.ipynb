## Challenge 1: Determinents of small and large singular matrices

1. Generate a 2x2 matrix with integers that have linear dependence
2. Generate random NxN matrices with linear dependance and compute determinant

```python
import numpy as np
from scipy import linalg as la
import sympy
```

```python
A_2 = np.random.randint(100, size = (2, 2))

A_2
```

```python
A_2[:, 1] = A_2[:, 0]

A_2
```

```python
np.linalg.det(A_2)
```

We can already see a small error in calculating the determinant here!

```python
A_30 = np.random.randint(100, size = (30, 30))
A_30[:, 1] = A_30[:, 0]
A_30
```

```python
np.linalg.det(A_30)
```

Yeesh. Not only is the determinant not even small for this 30x30 singular matrix, but the number generated is unrealistically large. This shows the pitfalls of using computers to calculate matrices. But there's absolutely no way you'd calculate this by hand either!
