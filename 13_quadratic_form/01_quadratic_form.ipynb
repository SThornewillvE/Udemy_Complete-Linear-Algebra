Consider that we can multiply any square matrix by a vector to give a vector, we can premultiply by that vector to then get the quadratic form.

$$w^TSw = \begin{bmatrix}x & y\end{bmatrix}
\begin{bmatrix}a & b \\ c & d \end{bmatrix}
\begin{bmatrix}x \\ y\end{bmatrix}$$

$$w^TSw = f(x) = ax^2 + (b+c)xy + dy^2$$

We can think of the vector x, y as being input coordinates to a matrix S and output a single value which becomes a surface.

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import *

%matplotlib inline
```

```python
A = np.random.randint(0, 100, (2,2))

A
```

```python
# Now let's see that A looks like in 100, 100 space
r = 1500 # resolution
surface = np.zeros(shape=(r,r))

d = r/2

for i in trange(r):
    for j in range(r):
        vec = np.array([i-d, j-d])
        energy = vec.dot(A).dot(vec.T)
        surface[i, j] = energy
```

```python
plt.matshow(surface)
plt.show()
```

This is one way to visualise the quadratic form of a matrix, but its kind of awkward and would be better in 3d. Let's try that.

Note that the eigenvectors will point the the directions of the surface. 

Now, here's an example of a matrix shown in class.

```python
r = 1000 # resolution
surface = np.zeros(shape=(r,r))
B = np.array([[4, 1],[1, 4]])


for i in trange(r):
    for j in range(r):
        vec = np.array([i, j])
        energy = vec.dot(B).dot(vec.T)
        surface[i, j] = energy
```

```python
# Find eigenvectors
e_val, e_vec = np.linalg.eig(B)

e_vec
```

```python
plt.matshow(surface)
plt.show()
```
